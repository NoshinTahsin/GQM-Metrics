```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dataset = pd.read_excel("EncodedData.xlsx")
#print(dataset.head)
dataset.dropna(inplace=True)
dataset.drop(columns=['Unnamed: 0'],inplace=True)
print(dataset.shape)
```
