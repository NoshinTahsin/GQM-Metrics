```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
dataset = pd.read_excel("EncodedDataWithoutDesignation.xlsx")
#print(dataset.head)
dataset.dropna(inplace=True)
dataset.drop(columns=['Unnamed: 0'],inplace=True)
print(dataset.shape)
```

```python
data=dataset[['Hobby Time','Family Time','Recreational Facilities','Outdoor Activities Frequency',
             'Average Daily Working Hour',
 'Decision Contribution',
 'Spontaneous Participation']]
cor=data.corr(method='spearman')
# setting the parameter values
annot = True
  
# plotting the heatmap
hm = sn.heatmap(data=cor,
                annot=annot)
  
# displaying the plotted heatmap
plt.show()
```
